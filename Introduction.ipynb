```python
def cleanWorld(word):
    return word.replace('.','').lower()

[cleanWorld(word) for word in myString.split()]
```

```python
def allPrimeUpTo(num):
    primes = [2]
    for number in range(3, num):
        sqrtNum = number**0.5
        for factor in primes:
            if number % factor ==0:
                #Not prime
                break
            if factor > sqrtNum:
                #It's prime!
                primes.append(number)
                break
    return primes
```

```python
print(allPrimeUpTo(100))
```

```python
print(allPrimeUpTo(1000))
```
